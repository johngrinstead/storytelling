```python
churn_mtm = churn[churn['Contract'] == 'Month-to-month']

churn_mtm.head()
```

```python
churn_mtm = churn[churn['Churn'] == 'Yes']

churn_mtm.head()
```

```python
sns.pairplot(data = churn_mtm)
```

```python
pd.crosstab(churn_mtm.StreamingTV, churn_mtm.StreamingMovies)
```

```python
churn_mtm['new_customer'] = churn_mtm.tenure < 6

churn_mtm.head()
```

```python
pd.crosstab(churn_mtm.PhoneService, churn_mtm.MultipleLines)

## almost all customer who churn have phone, of those only half have other lines of business
```

```python
pd.crosstab(churn_mtm.PaperlessBilling, churn_mtm.PaymentMethod)

## customers with paperless billing and pay by electronic check account for about half of all churn, why?
```

```python
pd.crosstab(churn_mtm.Dependents, churn_mtm.Partner)

## customers with no dependents, no partner, and are not seniors may be the largest churn group
```

```python
pd.crosstab(churn_mtm.Dependents, churn_mtm.new_customer)
```

```python
churn_mtm['live_alone'] = (churn_mtm.Dependents == 'No') & (churn_mtm.Partner == 'No')

churn_mtm.head()
```

```python
churn_mtm.groupby('live_alone').mean()
```

```python
churn_mtm.groupby('live_alone').customercount()
```

```python
pd.crosstab(churn_mtm.live_alone, churn_mtm.SeniorCitizen)

## half of churn is customers who live alone and are not seniors
```

```python
pd.crosstab(churn_mtm.live_alone, churn_mtm.PhoneService)

## people who live alone and have phone are a large group of customers who churn
```

```python
pd.crosstab(churn_mtm.live_alone, churn_mtm.MultipleLines)
```

```python
churn_mtm['alone_and_phone'] = (churn_mtm.live_alone == True) & (churn_mtm.PhoneService == 'Yes')

churn_mtm.head()
```

```python
pd.crosstab(churn_mtm.alone_and_phone, churn_mtm.MultipleLines)
```
